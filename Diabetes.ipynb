# Diabetes Analysis and Building a Machine Learning Model on Pima-Indians Dataset

Data mining and machine learning is helping medical professionals make diagnosis easier by bridging the gap between huge data sets and human knowledge. We can begin to apply machine learning techniques for classification in a dataset that describes a population that is under a high risk of the onset of diabetes.

Diabetes mellitus (DM), commonly known as diabetes, is a group of metabolic disorders characterized by a high blood sugar level over a prolonged period of time. Symptoms often include frequent urination, increased thirst, and increased appetite. If left untreated, diabetes can cause many complications. Acute complications can include diabetic ketoacidosis, hyperosmolar hyperglycemic state, or death. Serious long-term complications include cardiovascular disease, stroke, chronic kidney disease, foot ulcers, damage to the nerves, damage to the eyes and cognitive impairment.

The population for this study was the Pima Indian population near Phoenix, Arizona. The population has been under continuous study since 1965 by the National Institute of Diabetes and Digestive and Kidney Diseases because of its high incidence rate of diabetes.Data was collected between 1960s and 1980s. 

Mainly Data is collected from `768 Females` of Pima Indian Heritage

link for this Dataset is found in kaggle : https://www.kaggle.com/uciml/pima-indians-diabetes-database

By this Dataset we should be able to make better predictions on how likely a person is to suffer the onset of diabetes, and therefore act appropriately to help. We can start analyzing data and experimenting with algorithms that will help us study the onset of diabetes in Pima Indians.

##### We have 768 instances and the following 9 attributes : 
1. Pregnancies :  Number of times pregnant.
2. Glucose :Plasma glucose concentration a 2 hours in an oral glucose tolerance test. glucose in mg/dl. Above 160 is considered high.
3. BloodPressure :Diastolic blood pressure in mm Hg.  A normal reading would be any blood pressure below 120/80 mm Hg and above 90/60 mm Hg in an adult.
4. SkinThickness : Skin thickness is primarily determined by collagen content and is increased in insulin-dependent diabetes mellitus (IDDM).Triceps skin fold thickness in mm.
5. Insulin : 2-Hour serum insulin in mu U/ml. (units/ml)
6. BMI : Body mass index measured as weight in kg/(height in m)^2 (mass).
7. Diabetes pedigree function :It provided some data on diabetes mellitus history in relatives and the genetic relationship of those relatives to the patient. This measure of genetic influence gave us an idea of the hereditary risk one might have with the onset of diabetes mellitus. 
8. Age : Age in years
9. Outcome : 0-Non-Diabetic. 1-Diabetic.

```python
x= dia.iloc[:,:-1].values
y= dia.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

###### **Feature Scaling : Standard Scalar**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
clsfr = LogisticRegression(random_state=0)
clsfr.fit(x_train,y_train)
```

```python
y_pred=clsfr.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score
mat = confusion_matrix(y_test,y_pred)
ac_score =accuracy_score(y_test,y_pred)
f_score = f1_score(y_test,y_pred)
print(mat)
print(ac_score)
print(f_score)
```

#### Applying K-Fold Cross Validation

Inorder to get the best predicion results we use the K-Fold cross validation

K-fold cross-validation is used for experiment. consider k=10
This works in the following way:

Produce 10 equal sized data sets from given data.
Divide each set into two groups:  90% for training and 10% for testing.
Produce a classifier with an algorithm from 90% labeled data and apply that on the 10% testing data for set 1.
Continue for set 2 through 10
Average the performance of 10 classifiers produced from 10 equal sized (training and testing) sets

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = clsfr, X=x_train, y=y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

## Decision_tree_classification

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_pred_tree = classifier.predict(x_test)
cm = confusion_matrix(y_test, y_pred_tree)
ac=accuracy_score(y_test, y_pred_tree)
fs=f1_score(y_test,y_pred_tree)
print(cm)
print("Accuracy_score = {}".format(ac))
print("F1_score = {}".format(fs))
```

We transform few of the features by detectinhg the outliers.
1. Glucose
2. BloodPressure
3. Body Mass Index

```python
dia2=dia.copy()
```

```python
IQR_Glu=dia2.Glucose.quantile(0.75)-dia2.Glucose.quantile(0.25)
lower_bridge_1=dia2['Glucose'].quantile(0.25)-(IQR_Glu*2)
upper_bridge_1=dia2['Glucose'].quantile(0.75)+(IQR_Glu*2)
dia2.loc[dia2.Glucose<lower_bridge_1,"Glucose"]=lower_bridge_1
dia2.loc[dia2.Glucose>upper_bridge_1,"Glucose"]=upper_bridge_1
```

```python
IQR_bp=dia2.BloodPressure.quantile(0.75)-dia2.BloodPressure.quantile(0.25)
lower_bridge_2=dia2['BloodPressure'].quantile(0.25)-(IQR_bp*2)
upper_bridge_2=dia2['BloodPressure'].quantile(0.75)+(IQR_bp*2)
dia2.loc[dia2.BloodPressure<lower_bridge_2,"BloodPressure"]=lower_bridge_2
dia2.loc[dia2.BloodPressure>upper_bridge_2,"BloodPressure"]=upper_bridge_2
```

```python
IQR=dia2.BMI.quantile(0.75)-dia2.BMI.quantile(0.25)
lower_bridge=dia2['BMI'].quantile(0.25)-(IQR*2)
upper_bridge=dia2['BMI'].quantile(0.75)+(IQR*2)
dia2.loc[dia2.BMI<lower_bridge,"BMI"]=lower_bridge
dia2.loc[dia2.BMI>upper_bridge,"BMI"]=upper_bridge
```

```python
x1= dia2.iloc[:,:-1].values
y1= dia2.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train1,x_test1,y_train1,y_test1 = train_test_split(x1,y1,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc1 = StandardScaler()
x_train1=sc1.fit_transform(x_train1)
x_test1=sc1.transform(x_test1)
```

```python
from sklearn.linear_model import LogisticRegression
clsfr1 = LogisticRegression(random_state=0)
clsfr1.fit(x_train1,y_train1)
```

```python
y_pred1=clsfr1.predict(x_test1)
print(np.concatenate((y_pred1.reshape(len(y_pred1),1),y_test1.reshape(len(y_test1),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score
mat1 = confusion_matrix(y_test1,y_pred1)
ac_score1 =accuracy_score(y_test1,y_pred1)
f_score1 = f1_score(y_test1,y_pred1)
print(mat1)
print(ac_score1)
print(f_score1)
```
